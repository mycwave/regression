```python
train.drop(['UID'], axis=1, inplace=True)
test.drop(['UID'], axis=1, inplace=True)
train_data = train.drop(['target1','target2'],axis=1).values
train_target = train.loc[:,['target1','target2']].values
print(train_data.shape)
print(train_target.shape)
```

```python
from sklearn import cross_validation
#from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X_train, X_cv, y_train, y_cv = cross_validation.train_test_split(train_data, train_target, test_size = 0.1)



from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
scl.fit_transform(X_train, y_train)
scl.transform(X_cv)
scl.transform(test)


# Create a Randomized PCA model that takes two components
randomized_pca = RandomizedPCA(n_components=15)

# Fit and transform the data to the model
randomized_pca.fit_transform(X_train)
randomized_pca.fit(X_cv)
randomized_pca.fit(test)

reg = LinearRegression()

reg.fit(X_train,y_train)
pred_cv = reg.predict(X_cv)
print(pred_cv)

mse = mean_squared_error(y_cv, pred_cv)
print(mse)

#calculating r**2
print(reg.score(X_cv, y_cv))

test_data_pred=reg.predict(test)

#calculating coefficients 
#coeff = DataFrame(X_train)
#coeff['Coefficient Estimate'] = Series(reg.coef_)
#print(coeff['Coefficient Estimate'])

# Print out the data
#print(reduced_data_rpca.shape)
```
